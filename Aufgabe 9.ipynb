```python
%matplotlib inline
from IPython.display import display
from sympy import *
x = symbols('x')
i = symbols('i', integer=True)
init_printing()
```

```python
sstellen = [-2, -1, 0, 1, 2]
```

```python
L = [1, 1, 1, 1, 1]
for j in range(5):
  for k in range(5):
    if k != j:
      L[j] = L[j] * (x - sstellen[k])/(sstellen[j]-sstellen[k])
```

```python
p = plot(L[0], (x, -2.5, 2.5), show=False, line_color='r')
display(L[0])
lc = ['r', 'b', 'g', 'r', 'y']
for j in range(1, 5):
  display(L[j])
  p.extend(plot(L[j], (x, -2.5, 2.5), show=False, line_color=lc[j]))
```

```python
p.show()
```

```python
L_sum = 0
for i in range(5):
    L_sum += Abs(L[i])
    
display(L_sum)
```

```python
L_sum_strich = simplify(diff(L_sum, x))
L_sum_strich
```

```python
solve(L_sum_strich)
```
